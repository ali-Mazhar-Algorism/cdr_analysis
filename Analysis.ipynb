```python
def format_phone_number(number):
    # Remove any leading '+' if present
    number = number.lstrip('+')
    
    # Handle international numbers that start with '00'
    if number.startswith('00'):
        return number
    # Check if it already starts with the Pakistani country code '92' and has the correct length
    elif number.startswith('92') and len(number) == 12:
        return number
    # Handle local numbers starting with '03' and convert them to the correct format
    elif len(number) == 10 and number.startswith('03'):
        return '92' + number[1:]
    # Handle local numbers starting with '3' and convert them to the correct format
    elif len(number) == 11 and number.startswith('3'):
        return '92' + number
    # For other numbers, we will check if they seem like valid Pakistani numbers
    elif number.startswith('0') and len(number) == 11:
        return '92' + number[1:]
    elif number.startswith('92') and len(number) > 12:
        return '92' + number[-10:]
    else:
        # Keep the number as it is if it does not fit any of the above criteria
        return number
```

```python
def parse_site_info(info_string):
    # Handle case where the string is '?'
    if info_string == '?':
        return {
            'Address': '',
            'Latitude': '',
            'Longitude': '',
            'Connection code': '',
            'Internet': ''
        }
    
    # Split the string by the pipe '|' delimiter
    parts = info_string.split('|')
    
    # Extract latitude and longitude if available
    latitude = parts[1] if len(parts) > 1 else ''
    longitude = parts[2] if len(parts) > 2 else ''
    
    # Further split the first part to get connection code and address with internet info
    first_part = parts[0].split(', ')
    
    # Extract the connection code
    connection_code = first_part[0].strip() if len(first_part) > 0 else ''
    
    # Extract the internet type and address
    address_internet = first_part[1].split(' (') if len(first_part) > 1 else [first_part[0], '']
    address = address_internet[0].strip() if len(address_internet) > 0 else ''
    internet = address_internet[1].replace(')', '').strip() if len(address_internet) > 1 else ''
    
    # Create a dictionary to hold the parsed information
    parsed_info = {
        'Address': address,
        'Latitude': latitude,
        'Longitude': longitude,
        'Connection code': connection_code,
        'Internet': internet
    }
    
    return parsed_info
```

## Sample Chart

```python
import pydeck as pdk

DATA_URL = {
    "AIRPORTS": "https://raw.githubusercontent.com/visgl/deck.gl-data/master/examples/line/airports.json",
    "FLIGHT_PATHS": "https://raw.githubusercontent.com/visgl/deck.gl-data/master/examples/line/heathrow-flights.json",  # noqa
}

INITIAL_VIEW_STATE = pdk.ViewState(latitude=47.65, longitude=7, zoom=4.5, max_zoom=16, pitch=50, bearing=0)

# RGBA value generated in Javascript by deck.gl's Javascript expression parser
GET_COLOR_JS = [
    "255 * (1 - (start[2] / 10000) * 2)",
    "128 * (start[2] / 10000)",
    "255 * (start[2] / 10000)",
    "255 * (1 - (start[2] / 10000))",
]

scatterplot = pdk.Layer(
    "ScatterplotLayer",
    DATA_URL["AIRPORTS"],
    radius_scale=20,
    get_position="coordinates",
    get_fill_color=[255, 140, 0],
    get_radius=60,
    pickable=True,
)

line_layer = pdk.Layer(
    "LineLayer",
    DATA_URL["FLIGHT_PATHS"],
    get_source_position="start",
    get_target_position="end",
    get_color=GET_COLOR_JS,
    get_width=10,
    highlight_color=[255, 255, 0],
    picking_radius=10,
    auto_highlight=True,
    pickable=True,
)

layers = [scatterplot, line_layer]

r = pdk.Deck(layers=layers, initial_view_state=INITIAL_VIEW_STATE)
r.to_html("line_layer.html")
```

```python
df = raw_data

if 'Date & Time' in df.columns:
    df['Date & Time'] = pd.to_datetime(df['Date & Time'], errors='coerce')
    df = df.dropna(subset=['Date & Time'])  # Drop rows where 'Date & Time' is NaT
else:
    print("The 'Date & Time' column is missing or empty!")

# Check for valid Latitude and Longitude data
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
df = df.dropna(subset=['Latitude', 'Longitude'])
# Preview the processed data
# st.write(df[['Date & Time', 'Latitude', 'Longitude']].head())
# Create the start and end positions for each movement
df['start'] = df[['Longitude', 'Latitude']].shift(1).apply(list, axis=1)  # Previous location as list [Longitude, Latitude]
print(df['start'])
df = df.dropna()  # Drop the first row as it doesn't have a previous location
# Color based on time (using a gradient of blue to red)
df['time_norm'] = (df['Date & Time'] - df['Date & Time'].min()) / (df['Date & Time'].max() - df['Date & Time'].min())  # Normalize time
df['color'] = df['time_norm'].apply(lambda x: [255 * (1 - x), 128 * x, 255 * x, 200])  # RGBA gradient based on time
# Define the layers
scatterplot = pdk.Layer(
    "ScatterplotLayer",
    data=df,
    get_position=["Longitude", "Latitude"],
    get_fill_color=[255, 140, 0],  # Static color for points
    get_radius=100,
    pickable=True,
)
line_layer = pdk.Layer(
    "LineLayer",
    data=df,
    get_source_position="start",  # The previous point as a list
    get_target_position=["Longitude", "Latitude"],  # The current point
    get_color="color",  # Use the color calculated based on time
    get_width=5,
    auto_highlight=True,
    pickable=True,
)
# Initial map view
initial_view_state = pdk.ViewState(
    latitude=df['Latitude'].mean(),
    longitude=df['Longitude'].mean(),
    zoom=11,
    pitch=50,
)

    # Create the Deck.gl map
layers = [scatterplot, line_layer]
r = pdk.Deck(layers=layers, initial_view_state=initial_view_state, tooltip={"text": "{Date & Time}"})
r.to_html("line_layer.html")
```

### Using Plotly

```python




import plotly.express as px
import plotly.graph_objects as go


def line_tracking_chart(df):
    # Ensure the 'Date & Time' column is in datetime format and not empty

    df = df.sort_values(by='Date & Time')
    fig = go.Figure()

    # Extract data from the DataFrame
    latitudes = df['Latitude'].values
    longitudes = df['Longitude'].values
    times = df['Date & Time'].dt.strftime('%Y-%m-%d %H:%M:%S').values  # Time formatted for hover text
    
    color_data = df['Call Type'].apply(
        lambda x: [255, 0, 0, 255] if x == 'InComing' else  # Red
                ([0, 0, 255, 255] if x == 'Outgoing' else  # Blue
                ([0, 128, 0, 255] if x == 'Incoming SMS' else  # Green
                    ([255, 165, 0, 255] if x == 'Outgoing SMS' else [255, 255, 255, 160])))  # Orange
    ).tolist()
    
    df['color'] = color_data

    # Add scatter points for locations with lines connecting them
    fig.add_trace(go.Scattermapbox(
        lat=latitudes,
        lon=longitudes,
        mode='markers+lines',
        marker=dict(
            size=8,
            color=df['color'],  # Color column for color-coding (you can add this based on a condition)
            showscale=True,
            colorscale='Viridis',  # You can customize the colorscale
        ),
        line=dict(
            width=2,
            color='blue'  # Customize line color
        ),
        hoverinfo='text',
        hovertext=[f'Time: {t}' for t in times]  # Show 'Date & Time' on hover
    ))

    # Add arrows to indicate the movement direction
    for i in range(len(df) - 1):
        fig.add_trace(go.Scattermapbox(
            lat=[latitudes[i], latitudes[i + 1]],
            lon=[longitudes[i], longitudes[i + 1]],
            mode='lines',
            line=dict(width=2, color='red'),  # Color for the arrows (can be customized)
            marker=dict(size=4),
            hoverinfo='none'
        ))

    # Mapbox configuration (you can use open-street-map without a token)
    fig.update_layout(
        mapbox=dict(
            style="open-street-map",
            center=dict(lat=latitudes.mean(), lon=longitudes.mean()),  # Center the map
            zoom=12  # Adjust zoom level for better visualization
        ),
        margin={"r": 0, "t": 0, "l": 0, "b": 0}  # Remove margins around the map
    )

    fig.show()
```
